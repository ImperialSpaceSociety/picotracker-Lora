```python
#---------------------------------------
#
# Created: 30/09/2020
# Author: Medad Newman
#
#---------------------------------------

import re
from pathlib import Path
import json


line_selector = "([0-9]{4})-(0[1-9]|1[0-2])-(0[1-9]|[1-2][0-9]|3[0-1]) (2[0-3]|[01][0-9]):([0-5][0-9]):([0-5][0-9]),([0-9]{3}) (.*) b'(.*)'"


filename = "D:/picotracker-Lora/Current Tracker Prototype/firmware/Util scripts/Mapping/ICSPACE20-21.out"

raw_data = Path(filename).read_text()

lines = re.findall(line_selector, raw_data)

raw_data = [json.loads(i[8]) for i in lines]

with open("data_file.json", "w") as write_file:
    json.dump(raw_data, write_file)


```
